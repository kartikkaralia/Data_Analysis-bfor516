```python
import pandas as pd
import datetime
```

```python
# read in the data
weather = pd.read_csv('last_ten_alb.csv')
```

```python
weather.head()
```

```python
weather['timestamp'] = pd.to_datetime(weather['timestamp'])
```

```python
weather.head()
```

```python
weather.describe()
```

```python
annual_avg_temp = weather.groupby(pd.Grouper(key='timestamp', freq='Y')).mean()
annual_avg_temp
```

```python
monthly_avg_temp = weather.groupby(pd.Grouper(key='timestamp', freq='M')).mean()
monthly_avg_temp
```

```python
weather['mean_tmp']
```

```python
weather['mean_tmp'].plot()
```

```python
monthly_avg_temp.plot()
```
