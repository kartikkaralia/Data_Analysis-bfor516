# Lab 1 - Data Exploration
We will use the provided weather data to plot the mean temperature for each month in the dataset. To achieve the goal, `pandas` library would be used.

**Python Libraries**

First we will import the python libraries, pandas (a data management package) and datetime (a package to format dates).

```python
import pandas as pd
import datetime
```

**Import Data**

```python
# read file
weather = pd.read_csv('last_ten_alb.csv')
```

**View Data**

Using the .head() function from pandas, we can see the first 5 rows of the dataframe.

```python
# print first 5 rows
weather.head()
```

**Convert Dates**

Currently the dates are stored as strings, and we need to convert them to datetime format for further processing. For this, we will use the to_datetime() function.

```python
# convert dates
weather['timestamp']=pd.to_datetime(weather['timestamp'])
```

**Data about each Column**

To see details about each column, we will use the describe() function

```python
# column details
weather.describe(include='all', datetime_is_numeric=True)
```

**Average Temperature**

To get the average temperature, we can use the Grouper() function.

```python
# grouping by year
annual_avg_temp = weather.groupby(pd.Grouper(key='timestamp', freq='Y')).mean()
annual_avg_temp
```

Mean of monthly temperature can also be calculated by changing frequency period.

```python
monthly_avg_temp = weather.groupby(pd.Grouper(key='timestamp', freq='M')).mean()
monthly_avg_temp
```

**Mean Temperature of each month**

Although, we got the Average monthly temperatures, but they are on a daily basis. So, we need to group up data as monthly and then calculate the average.

```python
# group by month rater than intervals
monthwise_avg_temp = weather.groupby(weather['timestamp'].dt.month).mean()
monthwise_avg_temp
```

**Plotting the graph**

Data in a visual form is easy to read, understand and interpret from. So, we will plot a graph here for the month-wise average temperature. And for this purpose, we will use .plot() function.

```python
monthwise_avg_temp['mean_tmp'].plot(xlabel='Month', ylabel='Temp (in C)', title="Month-wise Avg Temp", )
```
